# SVM on internet advertising data 

#from http://archive.ics.uci.edu/ml/machine-learning-databases/internet_ads/

```python
import numpy as np
import pandas as pd

datafile = "./ad.data"
data = pd.read_csv(datafile, sep=',', header=None, low_memory=False)
data.head(11)
```

```python
#change missing values to NaN
```

```python
data.isnull().values.any()
```

```python
#replace all the empty cells with na
def toNum(cell):
    try:
        return np.float(cell)
    except:
        return np.nan

def seriestoNum(series):
    return series.apply(toNum)

train_data = data.iloc[0:, 0:-1].apply(seriestoNum)
train_data.head(11)
```

```python
train_data = train_data.dropna()
train_data.head(11)
```

```python
#Label data
```

```python
notad, ad = tuple(set(data.iloc[:, -1]))

def label(output):
    if (output == ad):
        return 1
    else:
        return 0
    
train_labels = data.iloc[train_data.index, -1].apply(label)
```

```python
#Train the model with Linear SVC
```

```python
from sklearn.svm import LinearSVC

clf = LinearSVC()
clf.fit(train_data[100:2300], train_labels[100:2300])
```

```python
#Test the data
```

```python
def predictedValue(cell):
    return clf.predict(cell.values.reshape(1, -1))
predicted_values = train_data.apply(predictedValue, axis=1)
```

```python
true_positives = 0
false_positives = 0
true_negetives = 0
false_negetives = 0

for predicted, actual in zip(predicted_values, train_labels):
    if (predicted[0] == actual == 1):
        true_positives += 1
    elif (predicted[0] == actual == 0):
        true_negetives += 1
    elif (predicted[0] == 1 and predicted[0] != actual):
        false_positives += 1
    else:
        false_negetives += 1
        
print("true_positives   =", true_positives)
print("false_positives  =", false_positives)
print("true_negetives   =", true_negetives)
print("false_negetives  =", false_negetives)
```

```python
precision = true_positives / ( true_positives + false_positives)
recall = true_positives/(true_positives + false_negetives)
accuracy= 2*precision*recall/(precision + recall)

print("precision  =", precision)
print("recall     =", recall)
print("accuracy   =", recall)
```
